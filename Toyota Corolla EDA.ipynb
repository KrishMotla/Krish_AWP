```python
#Rule 1 :
#if good correlation between input and output then no apply transformation
#if found skewness in given input column
#Rule 2: no apply transformation of -ve value of given column if found
#skewness
```

```python
#Before apply transformation on skewed columns ,
#check correlation between input and output
#here output price and rest are input
 #corr() inbuilt function , to find the correlation between
#2 variables
sns.heatmap(df_num.corr(),annot=True)
plt.show()
```

```python
#Apply Transformation on symbolic and normalized-losses because skewness found
#on both column and bed correlation between symbolic with output price and 
#normalized-losses with output price .
#but also check -ve value found in symbolic and normalized-losses column 
#or not 
#if found -ve value of symbolic and normalized-losses column  then no apply
#transformation of those columns
```

```python
ans=0 #suppose  no negative value
for v in df_num['symboling']:
    if v<0:
        ans=1 #negative value found
        print("Negative value found means no transformation apply ")
        break #exit from loop 
if ans==0:
    print("No negative value found")
```

```python
ans=0 #suppose  no negative value
for v in df_num['normalized-losses']:
    if v<0:
        ans=1 #negative value found
        print("Negative value found means no transformation apply ")
        break #exit from loop 
if ans==0:
    print("No negative value found")
```

```python
#clearly understand , apply transformation on normalized-losses column only
#to minimize the skewness 
#There are 2 types of transformation : -
#1. log()  and 2. sqrt() 
df_num['normalized-losses']=np.log(df_num['normalized-losses'])
```

```python
#Also check skew on df_num['normalized-losses'] columns
df_num['normalized-losses'].skew()
```
