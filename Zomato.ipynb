#### Dish Liked:

```python
df.isnull().sum()
```

```python
B = df[df['dish_liked'].isnull()]['cuisines'].unique()
for i in B:
    Z = df[df['cuisines'] == i]['dish_liked'].value_counts()
    if(len(Z) == 0):
        continue    
    else:
        Z = Z.index[0]
        df['dish_liked'][(df['dish_liked'].isnull()) & (df['cuisines'] == i)] = df['dish_liked'][(df['dish_liked'].isnull()) & (df['cuisines'] == i)].fillna(Z)
```

```python
df.isnull().sum()
```

```python
df[df['dish_liked'].isnull()]
```

```python
C = df[df['dish_liked'].isnull()]['rest_type'].unique()
for i in C:
    Z = df[df['rest_type'] == i]['dish_liked'].value_counts()
    if ( len(Z) == 0):
        continue
    else:
        Z = Z.index[0]
        df['dish_liked'][(df['dish_liked'].isnull()) & (df['rest_type'] == i)]  = df['dish_liked'][(df['dish_liked'].isnull()) & (df['rest_type'] == i)].fillna(Z)
```
